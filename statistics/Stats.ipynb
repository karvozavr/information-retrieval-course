```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import os
from subprocess import check_output
import tqdm
```

```python
sns.set(color_codes=True)
```

```python
directory = '/home/karvozavr/Documents/data/out'
text_dir = os.path.join(directory, 'text')
html_dir = os.path.join(directory, 'html')

def list_dir(directory):
    return sorted([f for f in os.listdir(directory) if os.path.isfile(os.path.join(directory, f))])
```

```python


def get_sizes(directory):
    return [os.stat(os.path.join(directory, f)).st_size for f in list_dir(directory)]

def get_wc(directory):
    return list(map(int, open('count').read().split()))
```

```python
text_sizes = np.array(get_sizes(text_dir))
html_sizes = np.array(get_sizes(html_dir))
```

```python
text_wc = np.array(get_wc(text_dir))
```

```python
plt.figure(figsize=(16,12))
print(text_wc.mean())
sns.distplot(text_wc[text_wc < 5000], axlabel='Size of text in words')
```

```python
#    1.4mb

plt.figure(figsize=(16,12))
print(text_sizes.mean())
sns.distplot(text_sizes[text_sizes < 40000], axlabel='Size of text in bytes')
```

```python
plt.figure(figsize=(16,12))

size_ratio = text_sizes / html_sizes

sns.distplot(size_ratio[size_ratio <= 1], axlabel='text to html ratio')

print(size_ratio.mean())
```
